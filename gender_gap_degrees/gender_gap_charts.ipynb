```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom=False, top=False, left=False, right=False)
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig, ax = plt.subplots(nrows=6, ncols=3)
fig.set_figheight(30)
fig.set_figwidth(20)

# First column
for n in range(6):
    # Set title
    ax[n][0].set_title(stem_cats[n])
    
    # Set lines for charts
    ax[n][0].plot(women_degrees['Year'], women_degrees[stem_cats[n]], c=cb_dark_blue, label='Women', linewidth=3)
    ax[n][0].plot(women_degrees['Year'], 100 - women_degrees[stem_cats[n]], c=cb_orange, label='Men', linewidth=3)
    
    # Clean labels
    ax[n][0].tick_params(bottom=False, top=False, right=False, left=False)
    ax[n][0].set_yticks([0, 100])
    ax[n][0].axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    if n != 5:
        ax[n][0].set_xticks([])
    for key, value in ax[n][0].spines.items():
        ax[n][0].spines[key].set_visible(False)
    
    # Set legend to lines
    if n == 0:
        ax[n][0].text(2005, 85, 'Women')
        ax[n][0].text(2007, 15, 'Men')
    if n == 5:
        ax[n][0].text(2005, 10, 'Women')
        ax[n][0].text(2007, 90, 'Men')              

# Second column
for n in range(5):
    # Set title
    ax[n][1].set_title(lib_arts_cats[n])
    
    # Set lines for charts
    ax[n][1].plot(women_degrees['Year'], women_degrees[lib_arts_cats[n]], c=cb_dark_blue, label='Women', linewidth=3)
    ax[n][1].plot(women_degrees['Year'], 100 - women_degrees[lib_arts_cats[n]], c=cb_orange, label='Men', linewidth=3)
    
    # Clean labels
    ax[n][1].tick_params(bottom=False, top=False, right=False, left=False)
    ax[n][1].set_yticks([0, 100])
    ax[n][1].axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    for key, value in ax[n][1].spines.items():
        ax[n][1].spines[key].set_visible(False)
    if n != 4:
        ax[n][1].set_xticks([])
    
    # Set legend to lines
    if n == 0:
        ax[n][1].text(2005, 80, 'Women')
        ax[n][1].text(2007, 15, 'Men')
    if n == 4:
        ax[n][1].text(2005, 40, 'Women')
        ax[n][1].text(2007, 60, 'Men')
        ax[n][1].axis('on')

# Clean sapce for last chart
ax[5][1].tick_params(bottom=False, top=False, right=False, left=False)
ax[5][1].axis('off')
for key, value in ax[5][1].spines.items():
        ax[5][1].spines[key].set_visible(False)

# Third column
for n in range(6):
    # Set title
    ax[n][2].set_title(other_cats[n])
    
    #Set lines for charts
    ax[n][2].plot(women_degrees['Year'], women_degrees[other_cats[n]], c=cb_dark_blue, label='Women', linewidth=3)
    ax[n][2].plot(women_degrees['Year'], 100 - women_degrees[other_cats[n]], c=cb_orange, label='Men', linewidth=3)
    
    # Clean labels
    ax[n][2].tick_params(bottom=False, top=False, right=False, left=False)
    ax[n][2].set_yticks([0, 100])
    ax[n][2].axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    for key, value in ax[n][2].spines.items():
        ax[n][2].spines[key].set_visible(False)
    if n != 5:
        ax[n][2].set_xticks([])
    
    # Set legend to lines
    if n == 0:
        ax[n][2].text(2005, 90, 'Women')
        ax[n][2].text(2007, 5, 'Men')
    if n == 5:
        ax[n][2].text(2005, 35, 'Women')
        ax[n][2].text(2007, 65, 'Men')

plt.savefig('Gender gap Exercise.png')
plt.show()
```
