```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'abtest',
       'vehicle_type', 'year_of_registration', 'gearbox', 'power_p_s', 'model',
       'odometer', 'month_of_registration', 'fuel_type', 'brand',
       'not_repaired_damage', 'ad_created', 'nr_of_pictures', 'postal_code',
       'last_seen']
```

```python
autos.head()
```

From now on, the columns have snakecase names, that easier to read and deal with Python code.

```python
autos.describe()
```

nr_of_pictures column have all columns with 0 value
power_p_s and year_of_registration have weird max numbers
price and odometer should be numbers instead of strings

```python
autos['price'] = (autos['price']
                  .str.replace(r'[$|,]', '')
                  .astype(float)
                 )
```

```python
autos['odometer'] = (autos['odometer']
                     .str.replace(r'[km|,]', '')
                     .astype(int)
                    )
autos.rename({'odometer':'odometer_km'}, axis=1, inplace=True)
```

In this analysis it is considered just modern model cars.
Despite of removing wrong data, cars older than 1900 will not be considered.

Before 1900, car's architecture were different from today. Thus, for eBay car section this type of product may have different consumers, prices and specifications.

This dataframe does not contain data input after 2016, so, car's model can't be for later then 2016.

```python
autos = autos.loc[autos['year_of_registration'].between(1900, 2016)]
```

```python
autos['year_of_registration'].value_counts(normalize=True, dropna= False).sort_values(ascending=False)
```

Half of the models have registration_year between 1999 - 2009.

```python
eligible_brands = (autos['brand']
                   .value_counts(normalize=True, dropna=False) > 0.005
                  ).value_counts()

eligible_brands
```

27 Brands have at least 5% share of the car's catalog were considered

```python
brands = (autos['brand']
          .value_counts(normalize=True)
          .sort_values(ascending=False).head(27)
         ).index
```

```python
brand_mean_prices = {}
brand_car_miles = {}

for b in brands:
    brand_mean_prices[b] = autos.loc[autos['brand'] == b, 'price'].mean()
    brand_car_miles[b] = autos.loc[autos['brand'] == b, 'odometer_km'].mean()

brand_mean_prices
```

Cars from 'other' brands have the most highest mean price, it can represent old brands or office brands. 

Despite this, the most expensive are: Audi, Mercedes, BMW and Mini.
On the other hand, Renault, Smart, Ford, Fiat and Citroen have the lowest mean prices

```python
bmp_series = pd.Series(brand_mean_prices)
bcm_series = pd.Series(brand_car_miles)
```

```python
b_df = pd.DataFrame(data={'mean_price':bmp_series,'mean_miles':bcm_series})
```

```python
b_df.sort_values(by='mean_miles', ascending=False)
```
